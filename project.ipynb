```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mll = pd.read_excel("FSI-2023-DOWNLOAD.xlsx") 
```

```python
top_10 = mll.sort_values(by="Total", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10, x="Total", y="Country", palette="Reds_d")
plt.title("Top 10 Most Fragile Countries (FSI 2023)")
plt.xlabel("FSI Score")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
numeric_mll = mll.select_dtypes(include='number').drop(columns=['Year'])
correlation = numeric_mll.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation, annot=True, cmap="coolwarm", square=True)
plt.title("Correlation Between FSI Indicators")
plt.tight_layout()
plt.show()
```

```python
def analyze_country(country_name):
    country_row = mll[mll['Country'].str.lower() == country_name.lower()]
    if country_row.empty:
        print(f"No data found for {country_name}")
        return

    row = country_row.iloc[0]
    print(f"\nCountry: {row['Country']}")
    print(f" Total Score: {row['Total']}")
    print(f" Rank: {row['Rank']}")
    
    # Show high-risk factors (> 9.0)
    print("\n⚠ High Risk Factors (score > 9.0):")
    for col in mll.columns[4:]:
        if row[col] > 9.0:
            print(f" - {col}: {row[col]}")

    print("\n Summary Categories:")
    print(f" - Economic: {row['E1: Economy']}, {row['E2: Economic Inequality']}")
    print(f" - Political: {row['P1: State Legitimacy']}, {row['P2: Public Services']}, {row['P3: Human Rights']}")
    print(f" - Security: {row['C1: Security Apparatus']}, {row['C2: Factionalized Elites']}, {row['X1: External Intervention']}")
```

```python
analyze_country("Yemen") 
```
